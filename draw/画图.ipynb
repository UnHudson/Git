```python
pic=pd.read_csv(r'E:/PycharmProjects/画图/housing.csv')
# pic=pic.astype(float)
pic['total_rooms']=pic['total_rooms'].astype(dtype='float')
```

```python
plt.figure(figsize=(20,15))
colors=np.random.rand(20640)
# 第一个图，经纬度下房间数
plt.subplot(2,2,1)
plt.scatter(x=pic.longitude,y=pic.latitude,s=pic.total_rooms/5,cmap='GnBu',alpha=0.3)
plt.title('房间数',fontsize=15)
# plt.scatter(x=pic.longitude,y=pic.latitude,label='总房间数',s=pic.total_rooms/500,c=colors,cmap='viridis',alpha=0.1)
#plt.colorbar()
# 第二个图，经纬度下 房间数和人口分布对比
plt.subplot(2,2,2)
plt.scatter(x=pic.longitude,y=pic.latitude,label='总房间数',s=pic.total_rooms/5,cmap='GnBu',alpha=0.3)
plt.scatter(x=pic.longitude,y=pic.latitude,s=pic.population/5,c='pink',alpha=0.2)
plt.title('房间数与人口关系',fontsize=15)
# 第三个图，位置和房间数（平均）的关系
plt.subplot(2,1,2)
plt.bar(x=pic.ocean_proximity[pic.ocean_proximity=='NEAR BAY'],height=pic.total_rooms[pic.ocean_proximity=='NEAR BAY'].mean(),color='pink',width=0.2)
plt.bar(x=pic.ocean_proximity[pic.ocean_proximity=='INLAND'],height=pic.total_rooms[pic.ocean_proximity=='INLAND'].mean(),color='lightpink',width=0.2)
plt.bar(x=pic.ocean_proximity[pic.ocean_proximity=='<1H OCEAN'],height=pic.total_rooms[pic.ocean_proximity=='<1H OCEAN'].mean(),color='palevioletred',width=0.2)
plt.title('不同区域平均房间数',fontsize=15)
plt.show()
```
